```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal
from scipy.fftpack import fft, fftfreq
from scipy.signal.windows import blackman
```

```python

b_fir = [1, 0.4, 0.2]  
a_fir = [1]             


w_fir, h_fir = signal.freqz(b_fir, a_fir)


plt.figure(figsize=(12, 6))
plt.subplot(2, 2, 1)
plt.plot(w_fir, 20 * np.log10(abs(h_fir)))
plt.title('FIR Frequency Response')
plt.ylabel('Magnitude (dB)')
plt.xlabel('Frequency [rad/sample]')


plt.subplot(2, 2, 2)
zeros_fir = np.roots(b_fir)
poles_fir = np.roots(a_fir)
plt.scatter(np.real(zeros_fir), np.imag(zeros_fir), marker='o', label='zero')
plt.scatter(np.real(poles_fir), np.imag(poles_fir), marker='x', label='pole')
plt.title('FIR Pole-Zero')
plt.xlabel('Real')
plt.ylabel('Im')
plt.legend()
plt.grid()

plt.tight_layout()
plt.show()
```

```python

b_iir = [1, 0.5]        
a_iir = [1, -0.9, 0.81] 


w_iir, h_iir = signal.freqz(b_iir, a_iir)


plt.subplot(2, 2, 1)
plt.plot(w_iir, 20 * np.log10(abs(h_iir)))
plt.title('IIR Response')
plt.ylabel('Magnitude (dB)')
plt.xlabel('Frequency [rad/sample]')


plt.subplot(2, 2, 2)
zeros_iir = np.roots(b_iir)
poles_iir = np.roots(a_iir)
plt.scatter(np.real(zeros_iir), np.imag(zeros_iir), marker='o', label='zero')
plt.scatter(np.real(poles_iir), np.imag(poles_iir), marker='x', label='pole')
plt.title('IIR Zero-Pole')
plt.xlabel('Real')
plt.ylabel('Im')
plt.legend()
plt.grid()

plt.tight_layout()
plt.show()
```

## 3. FFT

### a.

```python
def three_a():
    #     import numpy as np
    # import matplotlib.pyplot as plt
    # from scipy.fft import fft

    F = 2e6  
    Fs = 5e6  
    N = 64    

    
    t = np.arange(N) / Fs  

    
    x = np.cos(2 * np.pi * F * t)

    
    X = fft(x, N)

    
    frequencies = np.fft.fftfreq(N, 1/Fs)

    
    plt.figure(figsize=(10, 6))
    plt.stem(frequencies, np.abs(X), use_line_collection=True)
    plt.title('64-point DFT')
    plt.xlabel('Frequency (Hz)')
    plt.ylabel('Magnitude')
    plt.grid(True)
    plt.show()

three_a()
```

```python
def three_b():

    F1 = 200e6  
    F2 = 400e6
    Fs = 1e9  
    N = 64   

    t = np.arange(N) / Fs

    yt = np.cos(2*np.pi*F1*t) + np.cos(2*np.pi*F2*t)

    Y = fft(yt,N)
    

    frequencies = np.fft.fftfreq(N, 1/Fs)


    plt.figure(figsize=(10, 6))
    plt.stem(frequencies, np.abs(Y), use_line_collection=True)
    plt.title('64-point DFT')
    plt.xlabel('Frequency (Hz)')
    plt.ylabel('Magnitude')
    plt.grid(True)
    plt.show()

three_b()
```

```python
def three_d1():
    #     import numpy as np
    # import matplotlib.pyplot as plt
    # from scipy.fft import fft


    F = 2e6  
    Fs = 5e6  
    N = 64    

    window = blackman(N)

  
    t = np.arange(N) / Fs  

    
    x = np.cos(2 * np.pi * F * t)

    
    x_windowed = x * window

    
    X = fft(x_windowed, N)

   
    frequencies = np.fft.fftfreq(N, 1/Fs)

    
    plt.figure(figsize=(10, 6))
    plt.stem(frequencies, np.abs(X), use_line_collection=True)
    plt.title('64-point DFT')
    plt.xlabel('Frequency (Hz)')
    plt.ylabel('Magnitude')
    plt.grid(True)
    plt.show()

three_d1()
```
